```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots(figsize=(6,6))
ax.set_aspect('equal')

# Add main arrows
ax.quiver(0, 0, 1, 0, angles='xy', scale_units='xy', scale=1, color='black', width=0.005)
ax.quiver(0, 0, -1, 0, angles='xy', scale_units='xy', scale=1, color='black', width=0.005)
ax.quiver(0, 0, 0, 1, angles='xy', scale_units='xy', scale=1, color='black', width=0.005)
ax.quiver(0, 0, 0, -1, angles='xy', scale_units='xy', scale=1, color='black', width=0.005)

# Add angles and directions
ax.text(1.05, 0, 'upwind', fontsize=12, color='skyblue', ha='center')
ax.text(0, 1.05, 'crosswind', fontsize=12, color='skyblue', va='center')
ax.text(-0.8, -0.2, r'$\phi_i = 0^\circ$', fontsize=12, color='red', ha='center')
ax.quiver(-0.9, 0, 0.3, 0, angles='xy', scale_units='xy', scale=1, color='red', width=0.018)

# Scattering cone
theta = np.linspace(np.pi, 3/2*np.pi, 100)
r = np.ones_like(theta)
ax.plot(r * np.cos(theta), r * np.sin(theta), color='green', lw=2)
ax.text(-0.5, -0.5, 'scattering \n cone', fontsize=12, color='green', horizontalalignment='center')

# Backscattering arrow
ax.quiver(0, 0, -0.3, 0, angles='xy', scale_units='xy', scale=1, color='darkblue', width=0.009)
ax.text(-0.6, 0.05, 'Backscattering', fontsize=10, color='darkblue')

# Specular point
ax.quiver(0, 0, 0.3, 0, angles='xy', scale_units='xy', scale=1, color='gold', width=0.009)
ax.text(0.3, 0.05, 'specular', fontsize=10, color='gold')

# Wind vectors
wind_angles = [0.1*np.pi, 0.25 * np.pi, 0.35 * np.pi, 0.45 * np.pi]
for angle in wind_angles:
    ax.quiver(0 +np.cos(angle) , 0 + np.sin(angle), -np.cos(angle), -np.sin(angle),  angles='xy', scale_units='xy',ls = 'dashed', scale=1, color='skyblue',linewidth=.0, width=0.005)

# Labels for angles
ax.text(0, 0.7, r'$\phi_s = 90^\circ$', fontsize=10, color = 'green')
ax.text(-0.8, 0.2, r'$\phi_s = 180^\circ$', fontsize=10, color = 'green')
ax.text(0, -0.7, r'$\phi_s = 270^\circ$', fontsize=10, color = 'green')

# Set limits and hide ticks
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.set_xticks([])
ax.set_yticks([])

# Draw circles
circle1 = plt.Circle((0, 0), 1, color='green', fill=False, linestyle='--', lw=2)
ax.add_artist(circle1)

# Show plot
plt.show()
```
